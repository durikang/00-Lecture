```python
import pandas as pd

# 판다스 출력옵션 조절
pd.options.display.max_rows = 6 # 판다스 버전업에 따라 6행만 출력의 바뀐 코드

url = '/content/02_02_merge01.xlsx'
df1 = pd.read_excel(url, sheet_name=0, header=0)
df2 = pd.read_excel(url, sheet_name=1, header=0)


# df.merge(right, how='inner', on = None)
merged_df = df1.merge(df2, on='제품', how='left')

merged_df
```

복수의 열을 기준으로 vlookup을 하는 것

위 그림의 경우 업체와 음료 모두 기준 열로 삼아 두 데이터프레임을 병합하였다

```python
df3 = df1.merge(df2, how='left', on=['제품'])
df3.to_excel('merge_result01.xlsx',index=False) # to_excel('',index=False) index=False를통해 인덱스를 안갖고 오는 방법도 있음
```
