## 엑셀 파일을 데이터프레임으로 불러오는 함수

io
- 파일의 경로몀

sheet_name(인수는 문자열, 정수, 라스트/ 기본값은 0)
- 불러올 시트를 지정하는 인자

예)
- 지정하지 않을 때 : 첫번째 시트를 불러온다.
- 1: 2번째 시트를 불러온다
- "Sheet1": 문자열을 입력하면 해당 이름을 가진 시트를 불러온다. 여기서 "Sheet1"이라는 이름의 시트를 불러온다.

키(key)와 로케이션(location)
list = ['a','b','c']
로케이션 0   1   2
- 리스트는 앞에서부터 위치를 숫자로 부여한다.

list[0]
- 리스트는 로케이션을 이용해 원하는 개별 원소를 가져온다

dict = {'a':1,'b':2,'c':3}
- 딕셔너리는 키를 가지고 있다.

dict['a']
- 딕셔너리는 키로 벨류를 가져온다.

## 판다스는 키와 로케이션을 모두 사용할수 있을 대가 많은 것이 장점.

### header (인ㅇ수는 정수, 정수의 리스트 / 기본값은 0)
- columns를 지정하는 인자. 지정하지 않으면 맨 윗줄이 columns가 된다. 리스트로 지정하면 멀티인덱스인 columns가 된다.

### index_col(인수는 정수, 정수의 리스트/ 기본값은 None)
- index를 지정하는 인자. 지정하지 않으면 RangeIndex가 index로 부여된다.
리스트로 지정하면 멀티 인덱스인 index가 된다.
